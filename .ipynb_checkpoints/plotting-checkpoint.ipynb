## Plotting results

```python
import pandas as pd
```

```python
df = pd.read_csv('resultados.csv')
df.shape
```

### Separar colunas por heuristica.

```python
df_h1 = df.query('heuristic_function == "h1"')
df_h2 = df.query('heuristic_function == "h2"')
```

```python
df_merged = df_h1.merge(df_h2,
                        how='inner',
                        on='board',
                        suffixes=('_h1', '_h2'))
```

### Comparações heurísticas

```python
df_merged.head()
```

### Estados descobertos

```python
def diff_x1_to_x2(x1, x2):
    return x1 / x2
```

```python
df_estados_descobertos = df_merged[['board', 'estados_descobertos_h1', 'estados_descobertos_h2']].copy()
```

```python
df_estados_descobertos['estados_descobertos_h2_vezes_melhor'] = df_estados_descobertos[['estados_descobertos_h1', 'estados_descobertos_h2']].apply(lambda x: 
                                                                                                                                                   diff_x1_to_x2(x['estados_descobertos_h2'], 
                                                                                                                                                   x['estados_descobertos_h1']), 
                                                                                                                                                   axis='columns')
```

```python
df_estados_descobertos
```

```python
df_tempo_execucao = df_merged[['board', 'tempo_processamento_h1', 'tempo_processamento_h2']]
```

```python
df_tempo_execucao['tempo_processamento_h2_vezes_melhor'] = df_tempo_execucao[['tempo_processamento_h1', 'tempo_processamento_h2']].apply(lambda x: 
                                                                                                                                                   diff_x1_to_x2(x['tempo_processamento_h2'], 
                                                                                                                                                   x['tempo_processamento_h1']), 
                                                                                                                                                   axis='columns')
```
